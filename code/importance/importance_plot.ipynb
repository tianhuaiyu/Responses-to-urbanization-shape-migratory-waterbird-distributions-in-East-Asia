```python
import pandas as pd
import os 
import numpy as np
import matplotlib.pyplot as plt  
import matplotlib as mpl
%matplotlib inline 
import scipy.stats as ss 
import matplotlib.style as ms
import seaborn as sns
from sklearn import ensemble
#sns.set('whitegrid')
plt.rc('font',family='Arial')
plt.rcParams ['svg.fonttype'] ='none'
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'in'
```

```python
df = pd.read_csv(r'../../result/figure_data/variable_importance/west.csv',index_col=0)
df['Variable1']=['Elevation','Annual precipitation','Temperature','Distance-to-light','Distance-to-water', 'NDVI','Month',
 'Population density','Eastward wind components','Northward wind components','Precipitation', 
 'Cultivated and managed vegetation','Herbaceous vegetation','Barren','Shrubs','Mixed-Other trees',
'Evergreen deciduous needleleaf trees','Evergreen broadleaf trees', 'Deciduous broadleaf trees', 'Deciduous broadleaf trees' ]

df1 = pd.read_csv(r'../../result/figure_data/variable_importance/east.csv',index_col=0)
df1['Variable1']=['Annual precipitation','Elevation', 'Distance-to-light','NDVI','Temperature','Distance-to-water','Population density',
'Precipitation','Eastward wind components','Northward wind components','Month','Cultivated and managed vegetation',
'Barren','Herbaceous vegetation','Mixed-Other trees','Evergreen deciduous needleleaf trees', 'Deciduous broadleaf trees','Shrubs',
'Regularly flooded vegetation',  'Evergreen broadleaf trees',]
```

```python
df['Region']=['Western China']*len(df)
df1['Region']=['Eastern China']*len(df1)
df_merge = pd.concat([df,df1])
df_merge#.head(2)
```

```python
sns.set(style='white', font='Arial',font_scale=1.8)

fig,ax = plt.subplots(figsize = (12,9))
plt.setp(ax.get_xticklines(),visible=True)
plt.setp(ax.get_yticklines(),visible=True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xticks([0,0.1,0.2,0.3])
ax.set_xlim(0.0,0.305)
pal1 = sns.color_palette(['indianred','cornflowerblue'])
sns.barplot(x="Weight", y="Variable1", hue="Region", data=df_merge,ax=ax,palette=pal1)
ax.set_ylabel('')
ax.set_xlabel('Weight (Frequency of the feature)',fontdict={'size':24,'weight':'bold'},labelpad=7)
ax.legend(frameon=False)
sns.despine(top=False, right=False, left=False, bottom=False,ax=ax)
plt.savefig(r'../../result/figure/importance/importance.png',dpi=300,bbox_inches = 'tight')#_nospine
plt.savefig(r'../../result/figure/importance/importance.png',dpi=300,bbox_inches = 'tight')#_nospine
```
